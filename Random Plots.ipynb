```julia
x = linspace(200, 800, 20)
y = zeros(x)
y[1:17] = linspace(8e6, 3e5, 17)
y[18] = 1e5

scatter(x, y)

xlabel("Temperature", fontsize=25)
ylabel("Anisotropy", fontsize=25)
tick_params(labelsize=15)

plot([725, 725], [-2e6, 1e7], "--")

ylim(-0.1e7, 0.9e7)
xticks([])
yticks([])
text(750, 6e6, L"$T_C$", size=25)
```
